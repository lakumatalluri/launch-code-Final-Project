Prepare Your Dataframe

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# create a dataframe coviddf (covid_dataset.csv)
coviddf=pd.read_csv('covid_dataset.csv')
coviddf.head(5)
```

# Find Correlations

Use either pairplot or a heatmap or both to find the two variables with the strongest correlation in your dataset.
2 most correlated datapoints are 'Total Recovered','Total Deaths'

```python
plt.figure(figsize = (12,12))
sns.heatmap(coviddf.select_dtypes(['float64','int64']).corr(),annot =True)
plt.show()
```

## Create Your Model
Use the two columns with the strongest correlation to create and train your model. Make sure to print out the summary and plot the column values and the line produced by the model.

```python
# Creating X and y
X = coviddf['Total Recovered']
y = coviddf['Total Deaths']
# Splitting the varaibles as training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7,
                                                    test_size = 0.3, random_state = 100)
# Take a look at the train dataset
X_train
y_train
```

```python
# Importing Statsmodels.api library from Stamodel package
import statsmodels.api as sm
# Adding a constant to get an intercept
X_train_sm = sm.add_constant(X_train)
# Fitting the resgression line using 'OLS'
lr = sm.OLS(y_train, X_train_sm).fit()
# Printing the parameters
lr.params
# Performing a summary to list out all the different parameters of the regression line fitted
lr.summary()
```

```python
# Visualizing the regression line
plt.scatter(X_train, y_train)
plt.plot(X_train, 1260.2018+ 0.0100	*X_train, 'r')
plt.show()
```

# Error Terms
Finally, plot your error terms!

```python
# Predicting y_value using traingn data of X
y_train_pred = lr.predict(X_train_sm)
# Creating residuals from the y_train data and predicted y_data
res = (y_train - y_train_pred)
# Plotting the histogram using the residual values
fig = plt.figure()
sns.distplot(res, bins = 15)
plt.title('Error Terms', fontsize = 15)
plt.xlabel('y_train - y_train_pred', fontsize = 15)
plt.show()
```
